* Write a program that checks if a number is greater than 10 and prints "Greater than 10" if true.


```python
def greater_than_10(num):
  if num > 10:
    print('Number is greater than 10')
  elif num == 10:
    print('Number is equal to 10')
  else:
    print('Number less than 10')

```

* Write a program that checks if a number is even or odd and prints the result.


```python
def even_odd_checker(num):
  if num%2 == 0:
    print('Number is even')
  else:
    print('Number is odd')
```

* Write a program that checks if a number is positive, negative, or zero and prints the result.


```python
def number_identity(num):
  if num < 0:
    print('Negative')
  elif num > 0:
    print('Positive')
  else:
    print('zero')
```

* Write a program to find the sum of the first 10 natural numbers using a for loop.


```python
def sum_of_natural_numbers(n):
    sum = 0
    for i in range(n+1):
      sum = sum + i


    return sum
```

* Write a program that prints numbers from 1 to 10 but skips 5 using continue.


```python
for i in range(1,11):
  if i == 5:
    continue
  print(i)
```

* Write a program that prints numbers from 1 to 10 but stops when it reaches 7 using break.


```python
for i in range(1,11):
  if i ==7:
    break
  print(i)
```
