Problem 1: Calculate Mean (Average)

```python
def mean(num_list):
   total_sum = sum(num_list)
   result = total_sum / len(num_list)


   return result

mean([1,2,3])
```

Write a function dot_product(a, b) that takes two lists of numbers and returns their dot product without using NumPy.


```python
def dot_product(a,b):
    num = []
    for i in range(len(a)):
        prod = a[i] * b[i]
        num.append(prod)
    result = sum(num)
    return result

```
