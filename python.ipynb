problem: Two Sum
Write a function that takes a list of numbers and a target number. Return the indices of two numbers that add up to the target. (brute force approach, time complexity)

```python
def num_list_target(num_list, target):
  i = 0
  while i < len(num_list)-1:
    j = i + 1
    while j < len(num_list):
      if num_list[i] + num_list[j] == target:
        return [i,j]
      j = j +1
    i = i+1
  return[]

print(num_list_target([2,7,11,15], 9))
```

Problem: Find the Maximum Number
Write a function that takes a list of numbers and returns the largest number in the list. Don't use Python's built-in max() function - implement it yourself.

```python
def find_max_num(num_list):
    max_so_far = num_list[0]
    i = 1

    while i < len(num_list):
        if num_list[i] > max_so_far:
            max_so_far = num_list[i]
        i = i + 1

    return max_so_far

find_max_num([3,6,2])
```

Problem: Count Positive Numbers
Write a function that takes a list of numbers (positive, negative, and zero) and returns how many positive numbers are in the list.

```python
def return_positive_num(num_list):
  count = 0
  for i in range(len(num_list)):
    if num_list[i] > 0:
      count = count + 1
    i = i+1
  return count


return_positive_num([-1, 0, 1, 2, 3, -3, 6])
```

Next Problem: Sum of Even Numbers
Write a function that returns the sum of all even numbers in a list.

```python
def sum_even_num(num_list):
  sum = 0
  for i in range(len(num_list)):
    if num_list[i] % 2 == 0:
      sum = sum + num_list[i]
  i = i +1
  return sum


sum_even_num([2,4,3,5,9,10, -4])
```

Next Problem: Reverse a String
Write a function that takes a string and returns it reversed.


```python
def reverse_string(string):
    string_reversed = ''
    for i in range(len(string)):
     string_reversed = string[i] + string_reversed

    return string_reversed

reverse_string('cat')
```

Problem: Find the Smallest Number
Write a function that finds and returns the smallest number in a list (without using Python's built-in min() function).

```python
def find_smallest_num(num_list):
  smallest_num = num_list[0]
  for i in range(len(num_list)):
      if num_list[i] < smallest_num:
        smallest_num = num_list[i]
  return smallest_num


find_smallest_num([1, 7, 9, 0, 9, 5, 8, 6, -3, 5])
```
