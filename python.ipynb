problem: Two Sum
Write a function that takes a list of numbers and a target number. Return the indices of two numbers that add up to the target. (brute force approach, time complexity)

```python
def num_list_target(num_list, target):
  i = 0
  while i < len(num_list)-1:
    j = i + 1
    while j < len(num_list):
      if num_list[i] + num_list[j] == target:
        return [i,j]
      j = j +1
    i = i+1
  return[]

print(num_list_target([2,7,11,15], 9))
```

Problem: Find the Maximum Number
Write a function that takes a list of numbers and returns the largest number in the list. Don't use Python's built-in max() function - implement it yourself.

```python
def find_max_num(num_list):
    max_so_far = num_list[0]
    i = 1

    while i < len(num_list):
        if num_list[i] > max_so_far:
            max_so_far = num_list[i]
        i = i + 1

    return max_so_far

find_max_num([3,6,2])
```

Problem: Count Positive Numbers
Write a function that takes a list of numbers (positive, negative, and zero) and returns how many positive numbers are in the list.

```python
def return_positive_num(num_list):
  count = 0
  for i in range(len(num_list)):
    if num_list[i] > 0:
      count = count + 1
    i = i+1
  return count


return_positive_num([-1, 0, 1, 2, 3, -3, 6])
```

Problem: Find the Second Largest Number
Write a function that finds and returns the second largest number in a list. Assume the list has at least 2 numbers.


```python
def find_second_largest(num_list):
  largest_num = num_list[0]
  second_largest = num_list[0]

  for i in range(len(num_list)):
    if num_list[i] > largest_num:
      second_largest = largest_num
      largest_num = num_list[i]
    elif num_list[i] > second_largest:
      second_largest = num_list[i]

  return second_largest
```

Problem: Remove Duplicates
Write a function that takes a list and returns a new list with all duplicate values removed (keeping only the first occurrence of each number).

```python
def remove_duplicates(num_list):

  new_list = []
  for i in range(len(num_list)):
     if num_list[i] not in new_list:
        new_list.append(num_list[i])

  return new_list

remove_duplicates([2,2,3,3,4])
```

Problem: Palindrome Checker
Write a function that checks if a string is a palindrome. A palindrome reads the same forwards and backwards (like "racecar" or "madam").

```python
def palindrom_checker(string):

    for i in range(len(string)):
      if string[i] == string[len(string)-1-i]:
        return True
      else:
        return False

palindrom_checker('ahaha')
```

Problem: Merge Two Lists Alternately
Write a function that takes two lists and merges them by alternating elements from each list.

```python
def merge_list_alternatively(num_list_1, num_list_2):
    merged_list = []

    for i in range(len(num_list_1)):
      merged_list.append(num_list_1[i])
      merged_list.append(num_list_2[i])




    return merged_list


merge_list_alternatively([1,2,3],[4,5,6])
```

Next Problem: Find Common ElementsWrite a function that takes two lists and returns a new list containing only the elements that appear in BOTH lists (no duplicates in the result)

```python
def find_common_elements(num_list_1, num_list_2):

    common_elements = []

    for i in range(len(num_list_1)):
      if num_list_1[i] in num_list_2 and num_list_1[i] not in common_elements:
        common_elements.append(num_list_1[i])


    return common_elements


print(find_common_elements([1, 2, 2, 3], [2, 2, 4]))           # [2]
```

Problem: FizzBuzz
This is a classic programming problem! Write a function that takes a number n and returns a list from 1 to n, but:

Replace numbers divisible by 3 with "Fizz"

Replace numbers divisible by 5 with "Buzz"

Replace numbers divisible by both 3 and 5 with "FizzBuzz
"
Keep other numbers as they are

```python
def fizbuzz(n):
   new_list = []

   for i in range(1,n+1):
    if i%3==0 and i%5==0:
      new_list.append('fizzbuzz')
    elif i%3==0:
      new_list.append('fizz')
    elif i%5==0:
      new_list.append('buzz')
    else:
      new_list.append(i)


   return new_list

fizbuzz(10)
```

Problem: Group by Length
Write a function that takes a list of strings and groups them by their length. Return a dictionary where the keys are lengths and the values are lists of words with that length.


```python
def group_by_length(words):
    result = {}

    for i in range(len(words)):
        word_length = len(words[i])

        if word_length in result:
            # Key already exists, append word to existing list
            result[word_length].append(words[i])
        else:
            # Key doesn't exist, create new list with this word
            result[word_length] = [words[i]]

    return result


print(group_by_length(["hi", "cat", "dog", "apple", "to", "banana"]))
```
